```python
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d0 = pd.read_csv('mnist_train.csv')
```

```python
l = d0['label']
d = d0.drop('label', axis = 1)
```

```python
label = l.head(15000)
data = d.head(15000)

print('the shape of sample data = ', data.shape)
```

```python
# Data Preprocessing: standardizing the data

from sklearn. preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
data = standardized_data
labels = label[:]
```

```python
data.shape
```

```python
labels.shape
```

2 things are most important in the t-SNE method
1. perplexity : loosely held neighbours of a point
2. Iteration : Number of iteration to separate the data into categories

```python
model = TSNE(n_components = 2, random_state = 0)

# configuring the parameters
# number of components = 2
#default perplexity = 2
#default learning rate = 200 (step Size)
# manimum number of iterations for optimization = 1000

tsne_data = model.fit_transform(data)

# creating a new dataframe for plotting the results
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data = tsne_data, columns = ('Dim1', 'Dim2', 'labels'))

#ploting the result
sns.FacetGrid(tsne_df,hue = 'labels',size = 6).map(plt.scatter,'Dim1', 'Dim2', 'labels')
plt.show()
```

```python
# Increasing perplexity to 30 and number of iterations to 5000
model = TSNE(n_components = 2, random_state = 0,perplexity = 50, n_iter = 5000)

tsne_data = model.fit_transform(data)

# creating a new dataframe for plotting the results
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data = tsne_data, columns = ('Dim1', 'Dim2', 'labels'))

#ploting the result
sns.FacetGrid(tsne_df,hue = 'labels',size = 6).map(plt.scatter,'Dim1', 'Dim2', 'labels')
plt.show()
```

```python
model = TSNE(n_components = 2, random_state = 0,perplexity = 70, n_iter = 5000)

tsne_data = model.fit_transform(data)

# creating a new dataframe for plotting the results
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data = tsne_data, columns = ('Dim1', 'Dim2', 'labels'))

#ploting the result
sns.FacetGrid(tsne_df,hue = 'labels',size = 6).map(plt.scatter,'Dim1', 'Dim2', 'labels')
plt.show()
```
